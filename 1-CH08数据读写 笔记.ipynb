```python
#作业
# 利用百度POI小插件爬取一些poi数据，然后存成txt，再用python读取，编写成一个json形式（列表字典）：

#[{'name':'...', 'lng':..., 'lat':..., 'address':'...'},{...},...,]

import os
os.chdir('C:/Users/Administrator/Desktop/')        
f = open('jiuba.txt','r',encoding = 'gbk')         #先读取文件
m = []                    
n = 0                         #定义一个n的变量的目的是统计读取的数据总数
 
f.seek(0)                      #用readline函数来遍历文件里每一行信息
for line in f.readlines(): 
    n += 1                     #用这条语句来统计读取数据的总数
    st1 = line.split(':')      #  taxx:121.471271,31.229597,巨鹿路158号B1层B1号   先将一条信息用split函数，将‘：’前后的数据进行拆分
    name = st1[0]              #  拆分后得到的含有两个类数据的序列，名字数据以及经纬度及地址信息
    information = st1[1]       #  分别用序列的索引 st[]的方法 将名字信息进行复制，并且将剩下的信息作为一个整体序列进行下一步拆分
    st2 = information.split(',')   #再用split函数，将‘，’分隔的函数进行拆分，得到 经度 纬度 地址等信息的序列 
    lng = st2[0]
    lat = st2[1]
    add = st2[2].strip()       #再次通过序列索引的方法将得到的数据进行定义赋值
    data = [['name',name],['lng',lng],['lat',lat],['address',add]]    #因为最后的目的是生成字典类型的序列，而生成字典需要用到dict函数作用于列表类型数据
    m.append(dict(data))       #将每行读取的字典添加到列表中
    #print(dict(data))
print(m[:10])                  #输出前10条数据
print('\n总共转换数据%i条' %n)
```

```python
# writelines

f_w = open('test_w.txt','w',encoding = 'utf8')
lst = ['1','2','c','b']         #将数据输入文件一定要是字符串格式
for i in range(len(lst)):
    lst[i] = lst[i] + '\n'      #读入的每个数据进行换行
print(lst)
f_w.writelines(lst)
f_w.close()
```

```python
# 作业2
#两个列表[1~10],[a~j]，写入一个txt，变成以下格式

#1,a

#2,b

#3,c

#...

lst1 = list(range(1,11))
lst2 = ['a','b','c','d','e','f','g','h','i','j']

f = open('test_zy2.txt','w')
m = []       #我们首先要做的是将两组序列转化为：  ['1,a','2,b'......]   这种类型

for i in range(len(lst1)):
    lst = [str(lst1[i]),',',lst2[i]+'\n'] #需将lst1中的数转化为字符类型，再和lst2中的字符整合到一个字符串里
    #lst_ = str(lst1[i])+','+lst2[i]+'\n'  #将他们直接作为一个字符串连接也能运行出一样的结果
    m.append(lst)
    f.writelines(lst)

f.close()
print(m)
print('finished')
```

```python
'''
【课程8.4】  pickle模块的运用

pickle提供了一个简单的持久化功能。可以将对象以文件的形式存放在磁盘上。

python的pickle模块实现了基本的数据序列和反序列化

通过pickle模块的序列化操作我们能够将程序中运行的对象信息保存到文件中去，永久存储

通过pickle模块的反序列化操作，我们能够从文件中创建上一次程序保存的对象。

pickle.dump() / pickle.load()

'''
```

```python
# 存储：pickle.dump(obj, file, [,protocol])

import pickle
import os 

os.chdir('C:/Users/Administrator/Desktop/') 
data = {'a':[1,2,3,4],'b':1,'c':'hello world'}    # 创建一个字典变量data
print(data)

pic = open('data.pk1','wb')    # 以二进制来存储：rb, wb, wrb, ab
pickle.dump(m,pic)          # pickle.dump语句来保存数据
pic.close()
print('finished!')
```

```python
# 读取：pickle.load(file)

pic2 = open('C:/Users/Administrator/Desktop/data.pk1','rb')

data = pickle.load(pic2)    #pickle.load来读取文件
print(data)
```
